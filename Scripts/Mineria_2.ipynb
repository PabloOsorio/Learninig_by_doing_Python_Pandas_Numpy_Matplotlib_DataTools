```python
#Muestras
poblacion = df.copy()#Con eso creas otro objeto diferente(otra base da datos que no se vera afectada si modificas df)
poblacion.head(10)

```

```python
#Muestra sistemica
n=10

#Calcular en el intevalo
k=len(poblacion) // n

#Seleccionar punto de inicio
inicio=np.random.randint(0,k)

#Generar indices de la muestra
ims=np.arange(inicio,len(poblacion), step=k)
ims

#Obtener a muestra sistemica
muestra_sistemica=poblacion.iloc[ims]
muestra_sistemica.describe()
```

```python
#Muestra estratificada
#DICE EL PROFE QUE ESTA ES LA MEJOR

#Obtener los estratos
estratos=df.groupby("NivelEducacion")
estratos

#Seleccionar la muestra del % de cada estrato
muestra_estratificada = estratos.apply(lambda x: x.sample(frac=0.4,random_state=42))
muestra_estratificada =  muestra_estratificada.reset_index(drop=True)#Te ordena los numero que los enumera
muestra_estratificada
```

```python
dfp = df.copy()
dfp
```

```python
dfp["NivelEducacion"].unique()
#'Universidad', 'Bachillerato', 'Secundaria', 'Primaria']
nivel_educacion_cod={'Universidad':4, 'Bachillerato':3, 'Secundaria':2, 'Primaria':1}
dfp["NivelEducacionCod"]=dfp["NivelEducacion"].map(nivel_educacion_cod) #Tener cuidado de no borrar NivelEducacion
dfp
```

```python
dfp.drop(0,axis=0,inplace=True)#Eliminar renglon
dfp
```

```python
from sklearn.preprocessing import LabelEncoder
dfp["NivelEducacionLE"]=LabelEncoder().fit_transform(dfp.NivelEducacion)
dfp[["NivelEducacion","NivelEducacionCod","NivelEducacionLE"]]
```

```python
#One Hot Encoding = Binarizar
#Sirve para las categorias categoricas
one_hot=pd.get_dummies(dfp["NivelEducacion"],prefix="Ne")
one_hot
```

```python
#Unir los valores de one_hot con fdp
dfp=pd.concat([dfp, one_hot], axis=1)
dfp
```

```python
#Discretizar
dfp["EdadCategoria"]=dfp["Edad"].apply(lambda x: "AdultoMayor" if x >= 60 else "Adulto")
dfp[["Edad","EdadCategoria"]] #Mas de una columna
```

```python
dfp["AdultoMayor"]=dfp["EdadCategoria"].replace(["Adulto","AdultoMayor"],[0,1])
dfp[["Edad","EdadCategoria","AdultoMayor"]]
```

```python
def categorizar_edad(edad):
    if edad < 12:
        return 'Infante'
    elif 12 <= edad < 21:
        return 'Adolescente'
    elif 21 <= edad < 40:
        return 'Joven'
    elif 40 <= edad < 60:
        return 'Adulto'
    else:
        return 'Adulto Mayor'
```

```python
dfp["EdadCatAmplia"]=dfp["Edad"].apply(categorizar_edad)
dfp[["Edad","EdadCatAmplia"]]
```
