```python
#distribucin
plt.hist(df.Edad)
plt.title("Histograma de edad")
```

```python
#analisis de variables numericas con seaborn
for v in variables_numericas:
    plt.hist(df[v])
    plt.title(f'Histograma de {v}')
    plt.show()
```

```python
#analisis de variables numericas con seaborn
for v in variables_numericas:
    sns.histplot(df[v], kde=True, bins=20)
    plt.title(f'Histograma de {v}')
    plt.show()
```

```python
#analisis de variables categoricas
for v in variables_categoricas:
    print(f"frecuencia de {v}")
    print(df[v].value_counts())
    sns.countplot(x=v, data=df)
    plt.title(f"Distribucin de {v}")
    plt.show()
```

```python
#Anlisis de correlacion
print("Matrz de correlacin")
correlacion = df[["Ingresos", "Edad", "Casado", "Satisfaccion", "NumeroHijos"]].corr()
```

```python
sns.heatmap(correlacion, annot=True, cmap="coolwarm")
plt.title("Mapa de correlacin")
plt.show()
```

```python
#5 REALACION ENTRE VARIABLES
for v in variables_categoricas:
    sns.scatterplot(x="Edad", y="Ingresos", hue=v, data=df)
    plt.title(f"Relacion entre edad e ingresos por {v}")
    plt.show()
```

```python
#Relacion entre variables categoricas y Edad
for v in variables_categoricas:
    sns.boxplot(x=v, y="Edad", data = df, palette="deep")
    plt.title(f"edad por {v}")
    plt.show()
```

```python
#6 Analisis de variables categoricas ordinales
#Select NivelEducacion, AVG(Ingresos) FROM df Group by NivelEducacion
for v in variables_categoricas:
    print(df.groupby('NivelEducacion')['Ingresos'].mean())
    sns.barplot(x=v, y="Ingresos", data=df)
    plt.title(f"Ingresos promedio por {v}")
    plt.xlabel(v)
    plt.ylabel("Ingresos")
    plt.show()
```

```python
#7 Atipicos(Outliers)
for v in variables_numericas:
    sns.boxplot(x=df [v])
    plt.title(f"Boxplot de {v}")
    plt.show()
```
